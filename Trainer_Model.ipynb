```python
!pip install transformers datasets torch scikit-learn pandas numpy
import pandas as pd
from sklearn.model_selection import train_test_split
from datasets import Dataset

# ... (rest of your code) ...

# Assuming you have CSV files
train_df = pd.read_csv('/content/drive/MyDrive/Mini Project/TrainingDataset.csv')
test_df = pd.read_csv('/content/drive/MyDrive/Mini Project/TestDataset.csv')

# Convert to HuggingFace Dataset format
train_dataset = Dataset.from_pandas(train_df)
test_dataset = Dataset.from_pandas(test_df)

# If your data needs cleaning (example function)
def clean_text(text):
    # Add Telugu-specific cleaning if needed
    text = str(text).strip()
    # Remove special characters, URLs, etc.
    return text

train_dataset = train_dataset.map(lambda x: {'cleaned_text': clean_text(x['text'])})
test_dataset = test_dataset.map(lambda x: {'cleaned_text': clean_text(x['text'])})
```

```python
from transformers import AutoTokenizer

# Options for Telugu:
# - muril-base-cased (Google's multilingual BERT)
# - ai4bharat/indic-bert
# - xlm-roberta-base

model_name = "ai4bharat/indic-bert"  # Specifically works well for Indian languages
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_function(examples):
    return tokenizer(examples["cleaned_text"], padding="max_length", truncation=True, max_length=128)

tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_test = test_dataset.map(tokenize_function, batched=True)
```

```python
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

# Assuming sentiment labels are 0 (negative), 1 (neutral), 2 (positive)
num_labels = 3  # Adjust based on your dataset

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return {
        "accuracy": accuracy_score(labels, predictions),
        "f1": f1_score(labels, predictions, average="weighted"),
        "precision": precision_score(labels, predictions, average="weighted"),
        "recall": recall_score(labels, predictions, average="weighted")
    }
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    save_strategy="epoch",
    load_best_model_at_end=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    compute_metrics=compute_metrics,
)

trainer.train()
```

```python
# Evaluate
results = trainer.evaluate()
print(results)

# Make predictions
predictions = trainer.predict(tokenized_test)
preds = np.argmax(predictions.predictions, axis=-1)

# Confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

cm = confusion_matrix(tokenized_test['label'], preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
from wordcloud import WordCloud
from collections import defaultdict
import matplotlib.pyplot as plt
import requests
import os

def generate_sentiment_wordcloud(dataset, predictions):
    # Download a Telugu font (Noto Sans) if not present
    font_url = "https://github.com/googlefonts/noto-fonts/raw/main/hinted/ttf/NotoSansTelugu/NotoSansTelugu-Regular.ttf"
    font_path = "NotoSansTelugu-Regular.ttf"

    if not os.path.exists(font_path):
        try:
            response = requests.get(font_url, timeout=10)
            response.raise_for_status()
            with open(font_path, "wb") as f:
                f.write(response.content)
        except Exception as e:
            print(f"Font download failed: {e}")
            # Fallback to system font (may not support Telugu well)
            font_path = None

    # Group texts by sentiment
    sentiment_texts = defaultdict(str)
    for i, text in enumerate(dataset['cleaned_text']):
        sentiment = predictions[i]
        sentiment_texts[sentiment] += " " + str(text)  # Ensure text is string

    # Generate word clouds
    for sentiment, text in sentiment_texts.items():
        try:
            wc = WordCloud(
                width=800,
                height=400,
                background_color='white',
                font_path=font_path,
                collocations=False,
                regexp=r"[\w\']+",
                max_words=200,
                min_font_size=10,
                max_font_size=150,
                relative_scaling=0.5
            )

            wordcloud = wc.generate(text)

            plt.figure(figsize=(10, 5))
            plt.imshow(wordcloud, interpolation='bilinear')
            plt.title(f"Sentiment: {sentiment}", pad=20)
            plt.axis("off")
            plt.tight_layout()
            plt.show()

        except Exception as e:
            print(f"Error generating {sentiment} wordcloud: {str(e)}")
            continue

# Execute the function
generate_sentiment_wordcloud(test_dataset, preds)
```
